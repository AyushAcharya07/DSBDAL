```python
sns.histplot(x=df["SepalLengthCm"],bins=7)
mat.title("Sepal Length")
```

```python
sns.histplot(x=df["SepalWidthCm"],bins=7)
mat.title("Sepal Width")
```

```python
sns.histplot(x=df["PetalLengthCm"],bins=7)
mat.title("Petal Length")
```

```python
sns.histplot(x=df["PetalWidthCm"],bins=10) 
mat.title("Petal Width")
```

```python
sns.countplot(x='Species', data=df) 
mat.show()
```

```python
mat.figure(figsize=(10, 8))
sns.boxplot(data=df)
mat.title("Boxplots of Iris Dataset Features")
mat.show()
```

```python
sns.boxplot(x='SepalLengthCm', y='Species', data=df) 
mat.title("Boxplot of Sepal Length by Species") 
mat.show()
```

```python
sns.boxplot(x='SepalWidthCm', y='Species', data=df)
mat.title("Boxplot of Sepal Width by Species") 
mat.show()
```
