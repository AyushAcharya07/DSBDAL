#### Data Cleaning

```python
df.isna().sum()
```

```python
df["First Dose Administered"].fillna(value = df["First Dose Administered"].mean() , inplace=True)
df['First Dose Administered']=df['First Dose Administered'].astype(int)
```

```python
df['Second Dose Administered'].fillna(value = df["Second Dose Administered"].mean(),inplace=True)
df['Second Dose Administered']=df['Second Dose Administered'].astype(int)
```

```python
df["Total Doses Administered"].fillna(value = df["Total Doses Administered"].mean() , inplace=True)
```

```python
df["Total Individuals Vaccinated"].fillna(value = df["Total Individuals Vaccinated"].mean() , inplace=True)
```

```python
df['Male (Doses Administered)'].fillna(value=df['Male (Doses Administered)'].mean(),inplace=True)
```

```python
df['Female (Doses Administered)'].fillna(value=df['Female (Doses Administered)'].mean(),inplace=True)
```

```python
df['Male(Individuals Vaccinated)'].fillna(value=df['Male(Individuals Vaccinated)'].mean(),inplace=True)
df['Male(Individuals Vaccinated)']=df['Male(Individuals Vaccinated)'].astype(int)
```

```python
df['Female(Individuals Vaccinated)'].fillna(value=df['Female(Individuals Vaccinated)'].mean(),inplace=True)
df['Female(Individuals Vaccinated)']=df['Female(Individuals Vaccinated)'].astype(int)
```

```python
df.isna().sum()
```

### b. Number of persons state wise vaccinated for first dose in India

```python
first_dose=df.groupby("State")["First Dose Administered"].sum()
first_dose=first_dose.drop("India")
first_dose
```

#### Visualization

```python
first_dose.plot(kind='bar')
plt.title('Total First Doses Administered by State')
plt.xlabel('State')
plt.ylabel('First Dose Administered')
plt.show()
```

### c. Number of persons state wise vaccinated for second dose in India

```python
second_dose=df.groupby("State")["Second Dose Administered"].sum()
second_dose=second_dose.drop("India")
second_dose
```

#### Visualization

```python
second_dose.plot(kind='bar')
plt.title('Total Second Doses Administered by State')
plt.xlabel('State')
plt.ylabel('Second Dose Administered')
plt.show()
```

### d. Number of Males vaccinated

```python
male_doses=df['Male(Individuals Vaccinated)'].sum()
print("Total No. of Males Vaccinated in India are : ",male_doses)
```

### e. Number of females vaccinated

```python
female_doses=df['Female(Individuals Vaccinated)'].sum()
print("Total No. of Females Vaccinated in India are : ",female_doses)
```

#### Visualization

```python
labels = ['Male', 'Female']
sizes = [male_doses, female_doses]
colors = ['#ff9999','#66b3ff']
plt.figure(figsize=(8, 6)) 
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Vaccination by Gender in India')
plt.show()
```
