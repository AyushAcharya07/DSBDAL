```python
df["age"].fillna(df["age"].mean(),inplace=True) 
cnt=df['age'].isna().sum()
print(cnt)
```

```python
df["deck"].ffill(inplace=True) 
cnt2=df['deck'].isna().sum() 
print(cnt2)
```

```python
df["deck"].bfill(inplace = True)
cnt3=df['deck'].isna().sum()
print(cnt3)
```

```python
df["embarked"].bfill(inplace=True) 
cnt4=df['embarked'].isna().sum() 
print(cnt4)
```

```python
df["embark_town"].bfill(inplace = True) 
cnt5=df['embark_town'].isna().sum()  
print(cnt5)
```

```python
sex_count = df.groupby('sex')['sex'].count()
sex_count
```

```python
mat.figure(figsize=(7,7))
mat.title('Grouped by gender')
mat.pie(sex_count.values, labels=['male', 'female'], 
 autopct='%1.1f%%', textprops={'fontsize':13})
mat.show()
```
